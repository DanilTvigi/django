```python
import cv2  
import numpy as np  
from PIL import Image

def search(image, temp):
    if type(image) == type(""):
        img_rgb = cv2.imread(image,1)
    else:
        img_rgb = image
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)  
    template = cv2.imread(temp,0)  
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)   
    threshold = 0.8  
    loc = np.where( res >= threshold) 
    return img_rgb, loc  # 


def color_pixel(kletki, img, color):
    for key in kletki:
        sum_r, sum_g, sum_b = 0,0,0
        for i in range(0,50,2):
            for j in range(0,50,2):
                x, y = kletki[key]
                r, g, b = img.getpixel((x + i, y + j))
                sum_r += r
                sum_g += g
                sum_b += b
        color[key] = round((sum_r / 25) + (sum_g / 25) + (sum_b / 30))
    return color
```

```python
image = r'img/22.jpg'
# temp = r'RightUpB.jpg'

# Поиск границы поля
img_rgb, loc = search(image, 'template_edge.jpg') 

os_x = list()
os_y = list()

print(loc)

for pt in zip(*loc[::-1]):

    os_x.append(pt[0])
    os_y.append(pt[1])

x_min, y_min =  min(os_x) + 66, min(os_y) + 64
x_max, y_max =  max(os_x), max(os_y)

kletki = dict()
color = dict()
cords = list()

step_v = int((y_max - y_min) / 8)  #113.375
step_g = int((x_max - x_min) / 8)  #113.375

letter = ['A','B','C','D','E','F','G','H']
# letter = ['H','G','F','E','D','C','B','A']
for i in range(y_min,y_max-step_v, step_v):
    for j in range(x_min,x_max-step_g,step_g):
        img_rgb[i][j] = (0,0,255)
        cord = j ,i
        cords.append(cord)
chet = 0
for i in range(1,9):
    for j in letter:
        mark = j + str(i)
        kletki[mark] = cords[chet]
        color[mark] = ''
        chet += 1
    
img = Image.open(image)
color = color_pixel(kletki, img, color)
```

```python
def get_black_on_black(color):
    a1 = ['A','C','E','G']
    a11 = ['1']
    a2 = ['B','D','F','H']
    a22 = ['2']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для черных фигур на черном поле:\t",
    get_black_on_black(color)
)


#######################################################


def get_white_on_black(color):
    a1 = ['A','C','E','G']
    a11 = ['7']
    a2 = ['B','D','F','H']
    a22 = ['8']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для белых фигур на черном поле:\t",
    get_white_on_black(color)
)

#######################################################
#######################################################
#######################################################

def get_black_on_white(color):
    a1 = ['A','C','E','G']
    a11 = ['2']
    a2 = ['B','D','F','H']
    a22 = ['1']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для черных фигур на белом поле:\t",
    get_black_on_white(color)
)

def get_white_on_white(color):
    a1 = ['A','C','E','G']
    a11 = ['8']
    a2 = ['B','D','F','H']
    a22 = ['7']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для белых фигур на белых поле:\t\t",
    get_white_on_white(color)
)


#######################################################
#######################################################
#######################################################

def get_empty_black(color):
    a1 = ['A','C','E','G']
    a11 = ['3','5']
    a2 = ['B','D','F','H']
    a22 = ['4','6']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для пустых черных клеток:\t\t",
    get_empty_black(color)
)

def get_empty_white(color):
    a1 = ['A','C','E','G']
    a11 = ['4','6']
    a2 = ['B','D','F','H']
    a22 = ['3','5']

    tmp = []
    for key in color.keys():
        if key[0] in a1 and key[1] in a11:
            tmp.append(color[key])
        if key[0] in a2 and key[1] in a22:
            tmp.append(color[key])

    return [min(tmp), max(tmp)]

print(
    "Диапазон для пустых белых клеток:\t\t",
    get_empty_white(color)
)
```

```python
class board():
    def __init__(self):
        self.white_on_white = get_white_on_white(color) 
        self.white_on_black = get_white_on_black(color)
        self.black_on_white = get_black_on_white(color)
        self.black_on_black = get_black_on_black(color)
        self.empty_black = get_empty_black(color)
        self.empty_white = get_empty_white(color)

    def show(self):
        print(
            f"Белые на белом поле:  \t {self.white_on_white}", 
            f"Белые на черном поле: \t {self.white_on_black}", 
            f"Черные на белом поле: \t {self.black_on_white}", 
            f"Черные на черном поле:\t {self.black_on_black}", 
            f"Пустые белые клетки:  \t {self.empty_black}", 
            f"Пустые черные клетки: \t {self.empty_white}",
            sep="\n"
        )

myboard = board()
myboard.show()
```
